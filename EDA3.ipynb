```python
import pandas as pd
```

```python
df=pd.read_csv('tarantino.csv')
```

```python
df.tail(100)
```

```python
word_df = df[df['type'] == 'word'].groupby('movie')['word'].size().reset_index()
word_df
```

```python
death_df = df[df['type'] == 'death'].groupby('movie')['type'].size().reset_index()
death_df
```

Число проклятий, летальных исходов и их соотношение по фильмам 

```python
summary_df = word_df.merge(death_df, on='movie')
summary_df=summary_df.rename(columns={"type": "death"})
summary_df['ratio'] = summary_df['word'] / summary_df['death']
summary_df
```

Частота употребления конкретных проклятий

```python
word_df = df[df['type'] == 'word']
word_df.groupby('word').size().sort_values(ascending=False)
```

```python
word_df = word_df.sort_values(by=['movie', 'minutes_in']).reset_index(drop=True)
word_df['time_diff']=word_df.groupby("movie")["minutes_in"].diff()
word_df
```

```python
word_df=word_df.dropna()
word_df.describe()
```

Распределение времени между проклятиями

```python
import plotly.express as px
fig = px.histogram(word_df, x='time_diff', nbins=40, title='Распределение времени между проклятиями')

fig.update_layout(
    xaxis_title='Время между проклятиями (в секундах)',
    yaxis_title='Частота'
)

fig.show()
```
